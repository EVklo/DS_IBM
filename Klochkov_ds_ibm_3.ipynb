### This is a simple function to solve Josephus problem. If you want to know more, check [link (Wikipedia)](https://en.wikipedia.org/wiki/Josephus_problem)

```python
def josephus(items, k):
    dead_soldiers = [] #empty list of dead soldiers
    count = 1
    dead_list = [] #result sequence of dead soldiers
    while len(dead_soldiers) != len(items):
        for soldier in range(len(items)):
            if soldier not in dead_soldiers:
                if count == k:
                    dead_soldiers.append(soldier)
                    dead_list.append(items[soldier])
                    count = 1
                else:
                    count += 1
    return dead_list

josephus([1,2,3,4,5,6,7],3)==[3,6,2,7,5,1,4]
```
